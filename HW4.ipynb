Problem 0

Making a fake backgound to start.

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import h5py
import pandas as pd
import math

#this sets the size of the plot to something useful
plt.rcParams["figure.figsize"] = (10,5)
```

```python
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10 + np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
plt.plot(t,bkgd,'o')
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.title('Plot of the Background',fontsize=18)
plt.show
```

Problem 1.

We will now start looking at subsets of the data

1a)
Here is a scatter plot of the first 1000 points

```python
plt.plot(t[:1000],bkgd[:1000],'o',ms=8 )
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.title('Scatter plot of first 1000 data points',fontsize=18)

plt.show
```

1 b)
lets make a function so that we can quickly see some sections of the data

```python
def plotsect(start,lenth,x,y):
    plt.plot(x[start:start+lenth],y[start:start+lenth],'o' ,ms=5)
    plt.show
```

```python
plotsect(500,2000,t,bkgd)
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.title('Scatter plot of first 2000 points starting at 500',fontsize=18)
plt.show
```

1c) lets now only plot every 100th point 

```python
plt.plot(t,bkgd,'o',markevery=100,ms=8)
plt.title('Scatter plot of every 100th point of the backgound',fontsize=18)
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.show
```

Problem 2

2d)
now to make a 2D histagram 

```python
fig, ax = plt.subplots(1, 1)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
h = ax.hist2d(t,bkgd,bins=100, density=False)
plt.xlabel('t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.title('2D histagram of the background',fontsize=18)
plt.colorbar(h[3], ax=ax)
plt.show
```

2b) What we are plotting is the number of occurrences of a background measurements in a set time period. This is how we deal with overlapping points in a scatter plot. So the color is the number of point that we would see in a particular square if it was a scatter plot, so a purple square means 0 points there and a yellow square is a square with around 60 points of data

Problem 3

3a)
Scatter plot with folded data

```python
foldedT = t%(2*np.pi)
```

```python
plt.plot(foldedT,bkgd,'o',ms=3,markevery=1)
plt.title('Scatter plot of the folded backgound')
plt.xlabel('folded t',fontsize=18)
plt.ylabel('Background',fontsize=18)
```

3b) 2D histagram of the folded data

```python
fig, ax = plt.subplots(1, 1)
plt.rcParams["figure.figsize"] = (20,10)
h = ax.hist2d(foldedT,bkgd,bins=100, density=True);
plt.title('2D histagrm of the folded backgound')
plt.xlabel('folded t',fontsize=18)
plt.ylabel('Background',fontsize=18)
plt.colorbar(h[3], ax=ax)
plt.show
```
